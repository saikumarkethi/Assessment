**Task 2. Apply down-sampling method to reduce the sampling rate to 1 sample/minute**

Reading csv files and storing in a variable data1 data2 and data3.

```python
data1 = pd.read_csv('/content/detail.csv', parse_dates = ["Absolute Time"]) 
data2 =pd.read_csv('/content/detailVol.csv', parse_dates=['Realtime']) 
data3=pd.read_csv('/content/detailTemp.csv',parse_dates=['Realtime'])
```

Creating functions dsampleRealtime and dsampleABsolutetime dataset which contains Realtime and Absolute time respectively

```python
def dsampleRealtime(dataset):
    dataset.drop_duplicates(subset=['Realtime'], inplace=True)
    output = dataset.set_index('Realtime').resample('T').mean()
    return output
def dsampleAbsolutetime(dataset):
    dataset.drop_duplicates(subset=['Absolute Time'], inplace=True)
    output = dataset.set_index('Absolute Time').resample('T').mean()
    return output    
```

Storing output in 3 different variables for 3 different data fileds

```python
dtail=dsampleRealtime(data2)
dVol=dsampleRealtime(data3)
dTemp=dsampleAbsolutetime(data1)
```

Creating csv for each output variables


```python
dtail.to_csv("detailDownsampled.csv")
dVol.to_csv("detailVolDownsampled.csv")
dTemp.to_csv("detailTempDownsampled.csv")
```
